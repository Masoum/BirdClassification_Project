```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input
import matplotlib.pyplot as plt
from PIL import Image

```

```python

# Load my saved model
model = tf.keras.models.load_model('./VGG16_model.keras')

img_path = '../Data/images to predict/2.jpg'
img_path
```

```python
# Load the image, resize to 224x224 (VGG16 standard), and display it
img = Image.open(img_path).resize((224, 224))
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# Convert the image to a numpy array, add a batch dimension, and preprocess it
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = preprocess_input(img_array)
img_array
```

```python
# Run prediction
predictions = model.predict(img_array)

# finds the index of the highest value (the highest predicted probability) along each row (axis 1).
predicted_class_index = np.argmax(predictions, axis=1)[0]
print("Predicted class index:", predicted_class_index)
predictions
```
